```python
# int type data print

#정수
print(5)

#음수
print(-10)

#실수
print(3.14)
```

```python
# boolean type data print (True/False)

# False
print(5 > 10)

# True
print(5 < 10)

# not True = False, not False = True
print(not 20>10)
print(not 52<20)
```

<h2>3. 연산자 및 수식 (+, -, /, %, //, (, ), *, **, >, <, <=, >=, ==, =!, and, or)</h2>

```python
# 더하기 연산자 (addition)
print(32+21)

# 뻴셈 연산자 (subtraction)
print(4239-213)

# 곱셈 연산자 (multiplication)
print(372*2)

# 나눗셈 연산자 (division) 몫(소수점 포함), 나머지, 몫(소수점 제외)
print(321/32)
print(321%32)
print(321//32)

# 제곱 연산자 (square)
print(2**5)

# 일반 수식 연산자 작성의 경우에는 변수를 활용하여 다음과 같이 연산도 가능합 (모든 연산자가 사용 가능하나, 꼭 = 이전에 연산자를 두어아 오류 발생 안하니 주의)
number = 62
number += 32
print(number)
number -= 32
print(number)
number *= 2
print(number)
number /= 2
print(number)
number **= 2
print(number)

# 비교 연산자 (comparison) -> 결과는 True, False로 출력됨 (<=, >=와 같이 사용됩)
print(3<4)
print(3<1)
print(3==2)
print(3>=9)
print(8<=8)
print(8!=8)

# and(&) ,or(|) 연산자
print((3>0) and (2<4)) #-> True + True = True
print((3<0) | (3>9)) #-> False + False = False

# 수학 수식 연산자는 일반 수학 공식과 동일함
print(2+3*4)
print(2+(3*4))
print(2+3/4)
print(2+(3/4))
```
