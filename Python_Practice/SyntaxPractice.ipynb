<h2>3. 연산자 및 수식 (+, -, /, %, //, (, ), *, **, >, <, <=, >=, ==, =!, and, or)</h2>

```python
# 더하기 연산자 (addition)
print(32+21)

# 뻴셈 연산자 (subtraction)
print(4239-213)

# 곱셈 연산자 (multiplication)
print(372*2)

# 나눗셈 연산자 (division) 몫(소수점 포함), 나머지, 몫(소수점 제외)
print(321/32)
print(321%32)
print(321//32)

# 제곱 연산자 (square)
print(2**5)

# 일반 수식 연산자 작성의 경우에는 변수를 활용하여 다음과 같이 연산도 가능합 (모든 연산자가 사용 가능하나, 꼭 = 이전에 연산자를 두어아 오류 발생 안하니 주의)
number = 62
number += 32
print(number)
number -= 32
print(number)
number *= 2
print(number)
number /= 2
print(number)
number **= 2
print(number)

# 비교 연산자 (comparison) -> 결과는 True, False로 출력됨 (<=, >=와 같이 사용됩)
print(3<4)
print(3<1)
print(3==2)
print(3>=9)
print(8<=8)
print(8!=8)

# and(&) ,or(|) 연산자
print((3>0) and (2<4)) #-> True + True = True
print((3<0) | (3>9)) #-> False + False = False

# 수학 수식 연산자는 일반 수학 공식과 동일함
print(2+3*4)
print(2+(3*4))
print(2+3/4)
print(2+(3/4))
```

<h2>4. 숫자 관련 처리 함수 (abs, pow, max, min, round, math 라이브러리의 floor, ceil, sqrt, random 함수)</h2>

```python
# abs() : 절댓값
print(abs(-5))

#pow(a,b) : a에 대한 b제곱을 수행함, a**b랑 동일하지만 함수로 사용해야 하는 경우가 있으니 알아 둘 것
print(pow(4,2))

#max(a, b) : a,b 중 최대값을 출력함
print(max(4, 32))

#round() : 입력된 숫자의 소수점을 반올림하여 출력함
print(round(4.18))
print(round(5.99))

#math 라이브러리 import
from math import *

#floor() : 내림(소수점을 모두 버림)
print(floor(4.99))

#ceil() : 올림(소수점이 몇이든 1 높은 숫자로 변경)
print(ceil(16.32))

#sqrt() : 제곱근 값을 출력함(루르값의 결과)
print(sqrt(256))
```

```python
# random 함수 호출 방법
from random import *

# random() : 무작위 난수를 만들어냄. 단, 미 미지정시에는 0.0 ~ 1.0 사이의 값들 중에서 출력됨
print(random()) # 0.0 ~ 1.0 사이의 값
print(random()*10) # 소수점 값에 10을 곱함 (0 ~ 10 미만)
print(int(random()*10)) # int 형으로 변환하여 float 형의 소수점을 모두 날려버림

# random 함수를 활용해서 로또 번호를 랜덤하게 출력한다고 하면 다음과 같이 활용 가능함
list = []
for i in range(6):
    list.append(int(random()*45)+1) #random 값을 0~45이하에서 소수점 없는 정수형으로 만들어져서 list에 들어가도록
list.sort()
print(list)

# randrange() : 범위를 지정한 랜덤값 생성도 가능함
print(randrange(1, 46)) # 1 ~ 46 미만의 값 생성

# randint(a, b) : a~b 범위 내에서 a와 b를 포함한 값 중 랜덤값을 생성함
print(randint(1,45)) # 1 ~ 45의 값들 중 하나를 랜덤하게 출력함

# 따라서, 랜덤함수에는 지정된 값을 제외한 사이의 값 중 출력하는 random, 범위를 지정하여 사이의 값을 출력하는 randrange(), 범위를 지정하면 a,b로 주어진 값까지 포함시켜서 출력되는 randint()
# 가 있기에 추후에 수식이나 알고리즘 또는 DL, ML에서 원하는 수식을 만들어낼 경우에 필요하다면 적합한 함수를 골라서 사용하면 되겠다.
```

<h2>5. 문자열 처리 함수 (슬라이싱, len(), upper(), lower(), isupper(), replace(), index(), find() 등) </h2>

```python
# 변수에 문자열을 지정하여 출력하는 방법
lang = "Korean"
nation = "USA"
print("제 국적은 " + nation + "이며, 사용 가능한 외국어는 " + lang +"입니다.")

# 여러 줄의 문자열을 널어도 정상적으로 출력이 가능하다
sentence = '''
South Korea K-POP is definitly world class
my friend buy a lot of K-POP goods for 10 years
'''
print(sentence)

# 슬라이싱 (문자열 내에서 특정 범위 내의 원하는 문자열만 뽑아서 사용하는 방법)
ID_number = "001231-123234"
print("성별 : " + ID_number[7])
print("연 : " + ID_number[0:2]) #실 자리수 +1 까지 범위로 지정해야 포함됨
print("월 : " + ID_number[2:4])
print("일 : " + ID_number[4:6])
print("생년월일 : " + ID_number[:6]) # :(숫자)로 범위를 정하면 처음부터 지정된 인덱스 이전까지 문자열을 출력함
print("뒤 7자리 : " + ID_number[7:]) # (숫자):로 범위를 정하면 맨 뒷자리부터 지정된 인덱스까지의 문자열을 출력함
```

```python
# 문자열 처리 함수 
python = "Python is AmazingP" # 예시 문자열 준비

#lower() : 소문자로 모든 문자열을 변경
print(python.lower())

#upper() : 대문자로 모든 문자열을 변경
print(python.upper())

#isupper() : 해당 문자열이 대문자가 있는지 -> True/False
print(python[0].isupper())

#len() : 문자열의 길이를 출력함
print(len(python))

#replace("A", "B") : 문자열 A를 찾아서 B로 대체함 (문자, 단어 단위로 가능)
print(python.replace("n", "y"))

#index() : 문자열 내애서 원하는 문자의 위치를 찾을 수 있음
index = python.index("P")
print(index) #P의 문자열 위치를 index에 저장
index = python.index("P", index + 1) #처음 P를 찾은 위치 이후로 P가 있는지 검사하는 방법 -> 없을 경우에는 오류가 발생함
print(index) #두 번째 P의 위치를 출력

#find() : 원하는 문자의 위치를 알려주는 기능을 수행함
print(python.find("JAVA")) #찾을 경우 : 정확한 위치, 없을 경우 : -1

# index()와 find()를 보면 알 수 있듯이 둘 다 특정 문자나 문자열의 위치를 찾는 함수이지만
# index()는 없을 경우 오류를 출력하고 커널 자체를 멈춰버리고
# find()는 없을 경우 -1값을 return 하고 다음 명령을 이어나간다.
# 따라서, 이런 부분에서 차이가 있기 때문에 프로그래밍 하고자 하는 방식에 따라 유념하길 바란다.

#count() : 특정 문자의 반복 횟수를 문자열 내에서 검사하여 출력함
print(python.count("n")) #n의 갯수를 찾아서 출력함
```

<h2>6. 문자열 포맷</h2>

```python
# 기존의 문자열을 + , 를 통해서 이어 붙여서 사용하였다면, 해당 section에서는 다른 방법으로 가능함을 보여준다
print("a" + "b")
print("a", "b")

# %d : 정수형 값을 문자열 자리에 넣는 방법
print("올해는 %d년도 입니다." % 2025)
print("올해는 %d년도 이고, 내년은 %d년입니다" % (2025, 2026))

# %s : 문자열 값을 문자열 자리에 넣는 방법
print("올해 여행지 추천은 %s입니다" % "일본")

# %c : 문자를 문자열 자리에 넣는 방법
print("이번 학기 자료구조 수업의 학점은 %c입니다" % "A")

# {}, .format() -> format() 내에 원하는 데이터를 넣으면 자동으로 {} 내의 자리로 삽입된다.
print("내 올해 나이는 {}살이 됩니다. ".format(25))
print("올해는 {}년도 이고, 내년은 {}년입니다".format(2025, 2026))
print("올해는 {1}년도 이고, 내년은 {0}년입니다".format(2025, 2026)) #{} 내에 뒤에 넣고싶은 format의 문자열 순서를 임의로 변경하여 사용도 가능하다.
print("올해는 {1}년도 이고, 내년은 {1}년입니다".format(2025, 2026)) #{} 내에 같은 format 문자열의 인덱스를 넣으면 중복값이 들어간다
print("나는 {major}전공이며, 원하는 직업은 {job}입니다".format(major="컴퓨터공학과", job="개발자")) #이런 식으로 {}내에 변수를 적용하고 .format 내에서 변수 선언을 하면 적용이 가능하다.
# 해당 부분은 python 3.6 version 이상부터 사용 가능함
major = "Computer Science"
job = "Developer"
print(f"나는 {major}전공이며, 원하는 직업은 {job}입니다") #.format() 대신 변수를 외부에서 선언하고 출력문 앞에 f를 붙여서 {변수}를 적용하면 변수 내용이 문자열 내로 들어간다
```

<h2>8. 데이터 자료형 (List, Dict, Tuple, Set, 데이터 자료형 변경 방법)</h2>
<h3>1) List - index(), append(), insert(), pop(), count(), sort(), reverse(), clear(), extend()</h2>

```python
# List : []로 선언되며 다량의 데이터를 연속적인 공간에 모아서 자료형으로 사용하는 것이 특징이다

score = [90, 89, 52, 18, 99]
print(score)

# 이런 식으로 권장되지는 않지만, 문자열과 정수형 데이터가 함께 리스트에 들어가서 자료형을 이루는 것도 가능은 하다 (int, string, boolean 등 다양한 자료형 가능함)
list_ex = [90, "98", 11, "2"]
print(list_ex)

# index() : List 내부에 원하는 원소의 위치를 찾는 방법
list_ex.index("98")

# append() : 기준 리스트에 새로운 요소를 추가함
list_ex.append(92)
print(list_ex)

# insert(위치, 데이터) : List 내부에 원하는 위치에 삽입하는 방법
list_ex.insert(2, 1923)
print(list_ex)

# pop() : 리스트 맨 뒤에 요소부터 차례로 삭제하고 출력함
print(list_ex.pop())
print(list_ex)

# count() : List 내부에 같은 데이터가 몇개가 존재하는지 확인
list_ex.append(1923)
print(list_ex.count(1923))

# sort() : List 내부의 요소를 정렬
num_list = [2, 9, 43, 4, 3, 8, 12, 5]
num_list.sort()
print(num_list)

# reverse() : List 내부의 요소들의 순서를 앞뒤로 바꿔서 저장함
num_list.reverse()
print(num_list)

# clear() : List 내부의 요소를 모두 삭제함
num_list.clear()
print(num_list)

# extend() : 기존의 List에 다른 List를 확장시켜서 사용도 가능하다
original_list = [1, 2, 3, 4, 5]
extend_list = [6, 7, 8, 9, 10]
original_list.extend(extend_list)
print(original_list)
```

<h3> 4) Set - {}, set([List]), add(), remove() 합집합 : &, .intersection() 교집합 : union(), | 차집합 : -, .difference()</h3>

```python
# Set : Set은 Dict 자료형과 동일하게 {}를 통해서 선언하여 사용한다. Dict의 특징인 Key:Value 구조와는 다르게 Set은 Value 값만을 통해서 자료형을 구성한다
# Set 자료형의 특징은 1) 데이터의 중복 선언은 안됨, 2) 순서가 따로 없다는 것이 있다.

# {} : 해당 example에서도 볼 수 있듯이 set 자료형을 선언 시 중복된 데이터가 있어도 출력시에는 중복 데이터는 제거되어 나오는 것을 알 수 있다.
example_set = {1, 2, 3, 3, 4, 5, 5}
print(example_set)

# {}, set([List]) : Set은 수학적 개념으로는 집합이기 때문에 교집합, 합집합, 차집합의 개념도 구현이 가능하다.
team_roaster = {"박지성", "손홍민", "김민재"} #set은 {} 내에 이렇게 선언이 기본이지만
team2_roaster = set(["호날두", "손홍민"]) #[]리스트로 선언하고 후에 set()을 통해 set으로 변환도 가능하다

# &, .intersection() : 교집합에 해당하는 값을 출력
print(team_roaster & team2_roaster) 
print(team_roaster.intersection(team2_roaster))

# |, .union() : 합집합 (중복값을 제외한 모든 합집합의 원소를 출력함)
print(team_roaster | team2_roaster)
print(team_roaster.union())

# -, .difference() : 차집합 (기존 집합에서 교집합을 제외한 원소를 출력함)
print(team_roaster-team2_roaster)
print(team_roaster.difference(team2_roaster))

# add() : 기존 Set 자료형에 추가로 데이터를 삽입
team2_roaster.add("메시")
print(team2_roaster)

# remove() : 기존 Set 자료형 내부의 데이터를 삭제함
team2_roaster.remove("메시")
print(team2_roaster)
```

<h3>1) def(), return</h3>

```python
# def() : 함수를 정의하는 기능을 함
def open_account():
    print("새로운 계좌가 생성되었습니다.")

open_account()

# def(A, B) : 함수 내에 인자를 전달하는 방식으로도 정의가 가능함
def deposit(balance, money):
    balance += money
    print("입금이 완료되었습니다. 현재 잔액은 {0}원 입니다".format(balance))
    return balance

def withdraw(balance, money):
    if balance < money:
        print("잔액이 부족합니다. 현재 잔액 : {0}원".format(balance))
        return balance
    else:
        balance -= money
        print("출금이 완료되었습니다. 현재 잔액 : {0}원".format(balance))
        return balance
    
def withdraw_night(balance, money):
    commission = 1500
    if balance < money + commission:
        print("잔액이 부족합니다. 현재 잔액 : {0}원".format(balance))
        return balance
    else:
        balance -= money + commission
        print("야간 출금이 완료되었습니다. 현재 잔액 : {0}원, 야간 출금 수수료 : {1}원".format(balance, commission))
        return balance, commission #return 값도 다중으로 전달하여 변수에 전달 받을 수 있다.
 
balance = 0 #변수 생성
balance = deposit(balance, 59000) #59000원 입금
print(balance) #변수 내 반환된 값 출력

balance = withdraw(balance, 50000)
balance = withdraw(balance, 10000)
balance = withdraw(balance, 9000)
balance = deposit(balance, 10000)
balance = withdraw_night(balance, 2000)
```

<h2>12. 지역변수(local)와 전역변수(global) 선언 및 사용 방법</h2>

간단히 생각하면 지역변수는 함수의 호출 시 생성 및 사용 후 제거되는 것이 특징이며, 전역변수는 프로그램이 실행되는 동안 전역적으로 사용이 어디서든 가능한 변수라 생각하면 편하다

```python
# 지역변수
# 하단의 경우 checkpoint() 함수 실행 후 print()문의 경우 무조건 오류를 발생시키는데, 원인은 gun이라는 변수가 선언 및 값 할당도 되지 않았는데 사용하였기 때문이다.
# gun = 10이라 선언한 변수는 checkpoint() 함수 입장에서는 사용이 불가능한 지역변수이기 때문에 함수 내부에서 gun이라는 함수를 통한 연산이 불가능한 것이다
# gun = 10

# def checkpoint(soldiers):
#     gun = gun - soldiers
#     print("[함수 내] 남은 총의 갯수 : {0}".format(gun))
    
# print("전체 총 : {0}".format(gun))
# checkpoint(2) #gun의 갯수 -2
# print("남은 총 : {0}".format(gun))

# 전역변수
# 따라서, checkpoint() 함수의 이런 오류를 fix하기 위해서는 전역변수로서 함수 내에서 gun을 global로 선언하여 checkpoint() 내부에서도 사용이 가능하도록 해줘야 한다.
# 이렇게 선언할 경우 지역변수를 전역변수로 선언하여 다른 함수들 내부에서도 값을 통해서 연산이 가능하도록 사용이 가능하다.
gun = 10

def checkpoint(soldiers):
    global gun # gun 변수의 전역변수 선언
    gun = gun - soldiers
    print("[함수 내] 남은 총의 갯수 : {0}".format(gun))
    
print("전체 총 : {0}".format(gun))
checkpoint(2) #gun의 갯수 -2
print("남은 총 : {0}".format(gun))
    
# 위와 같은 전역변수를 통한 사용을 하기 싫다면, 지역변수(local)로 선언된 gun 값을 인자로 전달받아서 함수 내부에서 연산이 가능하도록 하고
# 값을 return 값으로 반환을 받아서 출력하는 방법도 있다.
def checkpoint_ret(gun, soldiers):
    gun = gun - soldiers
    print("[함수 내] 남은 총 : {0}".format(gun))
    return gun

gun = checkpoint_ret(gun, 2) #gun의 갯수 -2
print("남은 총 : {0}".format(gun))
```

class가 요구하는 인자와 변수명에 객체를 생성하면 훨씬 간편하게 다중의 객체를 control이 가능하기 때문에 모든 프로그래밍 언어에서 사용되는 기능 중 하나로 알려져 있다

```python
# Terran의 marine 구현

marine_name = "marine" #유닛 이름
marine_hp = 40 # 체력
marine_damage = 5 # 공격력

print("{0} 유닛이 생성되었습니다.".format(marine_name))
print("체력 {0}, 공격력 {1}\n".format(marine_hp, marine_damage))

# Terran의 tank 구현

tank_name = "tank"
tank_hp = 150
tank_damage = 35 

print("{0} 유닛이 생성되었습니다.".format(tank_name))
print("체력 {0}, 공격력 {1}\n".format(tank_hp, tank_damage))

def attack(name, location, damage):
    print("{0} : {1} 방향으로 적군을 공격합니다. [공격력 {2}]".format(name, location, damage))
    
attack(marine_name, "1시", marine_damage)
attack(tank_name, "1시", tank_damage)

# 이런 과정에서는 하나씩 유닛을 만들어보면 문제는 안되는 것처럼 보이지만, 유닛이 다수로 많아진다면 생성에 힘든 과정이 너무 많기 때문에
# class 기능을 통해서 선언하여 한번의 클래스 호출로 다수의 유닛을 생성 가능하도록 만드는 방법으로 사용한다.

class Unit: #class 생성
    def __init__(self, name, hp, damage): #인자 생성
        self.name = name
        self.hp = hp
        self.damage = damage
        # 함수 호출 시 실행문
        print("{0} 유닛이 생성되었습니다.".format(self.name))
        print("체력 {0}, 공격력 {1}\n".format(self.hp, self.damage))
        
marine1 = Unit("marine", 40, 5) #marine 객체 생성
marine2 = Unit("marine", 40, 5) #marine 객체 생성
tank = Unit("tank", 150, 35) #tank 객체 생성
```

<h3>2) __init__이란? -> 클래스 내부에 __init__(self, ): 형태로 선언</h3>

__init__은 클래스를 통해서 만들어지는 객체들의 인스턴스로 인자를 전달하도록 선언하는 부분을 담당한다.

따라서, 객체를 생성할 경우 __init__에 정해진 인자값을 전달하며 선언을 해야 클래스에 해당하는 객체 생성이 가능하기에 이를 주의하여 사용해야 한다.

```python
class Unit: #class 생성
    def __init__(self, name, hp, damage): #인자 생성
        self.name = name
        self.hp = hp
        self.damage = damage
        # 함수 호출 시 실행문
        print("{0} 유닛이 생성되었습니다.".format(self.name))
        print("체력 {0}, 공격력 {1}\n".format(self.hp, self.damage))
```

<h3>3) 멤버변수란?</h3>

멤버변수는 하단 코드에서 self.name, self.hp, self.damage로 선언되는 변수를 의미하는데 객체를 생성하였다면 

객체변수명.변수명으로 호출하여 클래스 내부의 멤버변수의 값을 외부에서도 사용이 가능하다 (marine.name, marine,hp, marine.damage)

```python
class Unit: #class 생성
    def __init__(self, name, hp, damage): #인자 생성
        self.name = name #멤버변수
        self.hp = hp #멤버변수
        self.damage = damage #멤버변수
        # 함수 호출 시 실행문
        print("{0} 유닛이 생성되었습니다.".format(self.name))
        print("체력 {0}, 공격력 {1}\n".format(self.hp, self.damage))
        
# 객체 생성 후 멤버변수 활용 예시
wraith1 = Unit("wraith", 80, 5)
print("유닛 이름 : {0}, 공격력 : {1}".format(wraith1.name, wraith1.damage))

# 기존 멤버변수에 없던 값을 외부에서 객체에 추가하여 변수를 할당하여 사용이 가능하다
wraith2 = Unit("stollen wraith", 80, 5)
wraith2.clocking = True #뺏긴 상태 활성화 변수 추가 (이런 방식으로 기존 클래스 내부에서 생성된 멤버변수가 아닌 값을 추가로 외부에서 할당하여 사용도 가능하다.)

if wraith2.clocking == True:
    print("{0}은 현재 클로킹 상태로 제어가 불가합니다.".format(wraith2.name))
```

<h3>4) 메소드란?</h3>

객체 내부에 함수를 정의하여 객체를 이용하여 클래스 내부에 선언된 함수를 호출한다는 점이 특징이다.

함수는 이름 자체로 호출이 가능한 반면 메소드는 클래스 내부에 선언되어 있기 때문에 클래스명.메소드명으로 선언하여 사용해야 한다. (ex) marine.attack())

```python
class Unit: #class 생성
    def __init__(self, name, hp, damage): #인자 생성
        self.name = name #멤버변수
        self.hp = hp #멤버변수
        self.damage = damage #멤버변수
        # 함수 호출 시 실행문
        print("{0} 유닛이 생성되었습니다.".format(self.name))
        print("체력 {0}, 공격력 {1}\n".format(self.hp, self.damage))
        
class AttackUnit:
    def __init__(self, name, hp, damage): #인자 생성
        self.name = name #멤버변수
        self.hp = hp #멤버변수
        self.damage = damage #멤버변수
    
    def attack(self, location):
        print("{0} : {1} 방향으로 적군을 공격합니다. [공격력 {2}]".format(self.name, location, self.damage))
        
    def damaged(self, damage):
        print("{0} : {1} 데미지를 입었습니다.".format(self.name, damage))
        self.hp -= damage
        print("{0} : 현재 체력은 {1} 입니다".format(self.name, self.hp))
        if self.hp <= 0:
            print("{0} : 해당 유닛이 파괴되었습니다.".format(self.name))
            
# 파이어벳 생성 : 공격 유닛, 화염방사기
firebat1 = AttackUnit("파이어벳", 50, 16)
firebat1.attack("5시")

# 공격 2번을 가정
firebat1.damaged(25)
firebat1.damaged(25)
```

<h2>17. 상속과 다중상속</h2>

```python
# 상속의 경우, 기존에 존재하는 클래스를 기반으로 새로운 클래스를 정의할 때 기존 클래스의 멤버변수를 사용하고자 한다면 
# class 클래스명(상속할클래스명): 으로 선언하여 클래스 내부에서 상속 받은 클래스의 멤버변수를 상속받은클래스명.__init__(self, 변수명, 변수명)으로
# 가져와서 사용해야 한다.

# 다중상속의 경우에는 하나 이상의 클래스에서 상속을 다중으로 받는다는 의미로 여러 클래스를 상속받아 필요한 멤버 변수를 상속받은클래스명.__init__(self, 변수명, 변수명)으로
# 각각 def __init__(): 내부에 초기화를 진행하여 따로 정의없이 상속만으로 사용이 가능하다는 점에서 이점이 크다.

# 일반 유닛
class Unit:
    def __init__(self, name, hp):
        self.name = name
        self.hp = hp

# 공격 유닛    
class AttackUnit(Unit):
    def __init__(self, name, hp, damage):
        Unit.__init__(self, name, hp) #상속받은 클래스의 멤버변수를 호출
        self.damage = damage
    
    def attack(self, location):
        print("{0} : {1} 방향으로 적군을 공격합니다. [공격력 {2}]".format(self.name, location, self.damage))
        
    def damaged(self, damage):
        print("{0} : {1} 데미지를 입었습니다.".format(self.name, damage))
        self.hp -= damage
        print("{0} : 현재 체력은 {1} 입니다".format(self.name, self.hp))
        if self.hp <= 0:
            print("{0} : 해당 유닛이 파괴되었습니다.".format(self.name))
            
firebat1 = AttackUnit("파이어뱃", 50, 16)
firebat1.attack("5시")

firebat1.damaged(25)
firebat1.damaged(25)
            
# 수송기 유닛 (공격기능은 없음)
class FlyableUnit:
    def __init__(self, flying_speed):
        self.flying_speed = flying_speed
        
    def fly(self, name, location):
        print("{0} : {1} 방향으로 날아가고 있습니다. [속도 {2}]".format(name, location, self.flying_speed))
        
# 공격기 유닛
class FlyableAttackUnit(AttackUnit, FlyableUnit): #다중 상속
    def __init__(self, name, hp, damage, flying_speed):
        AttackUnit.__init__(self, name, hp, damage) #상속 받는 클래스의 멤버 변수를 초기화 진행
        FlyableUnit.__init__(self, flying_speed) #상속 받는 클래스의 멤버 변수를 초기화 진행
     
# 공중 유닛 생성
valkyrie = FlyableAttackUnit("발키리", 200, 6, 5)
valkyrie.fly(valkyrie.name, "3시")
```

<h3>1) 메소드 오버라이딩 (부모 클래스 <- 자식 클래스)</h3>

메소드 오버라이딩이란, 부모 클래스에는 없지만 자식 클래스에 있는 메소드를 역으로 부모 클래스에서 끌어와(오버라이딩)서 사용하는 방법을 의미한다.

```python
class Unit:
    def __init__(self, name, hp, speed):
        self.name = name
        self.hp = hp
        self.speed = speed #기존에 없던 speed 메소드를 추가 선언
        
    def move(self, location):
        print("[지상 유닛 이동]")
        print("{0} : {1} 방향으로 이동합니다. [속도 {2}]".format(self.name, location, self.speed))
        
class AttackUnit(Unit):
    def __init__(self, name, hp, damage, speed):
        Unit.__init__(self, name, hp, speed) # speed를 추가로 상속받아서 초기화 진행
        self.damage = damage
    
    def attack(self, location):
        print("{0} : {1} 방향으로 적군을 공격합니다. [공격력 {2}]".format(self.name, location, self.damage))
        
    def damaged(self, damage):
        print("{0} : {1} 데미지를 입었습니다.".format(self.name, damage))
        self.hp -= damage
        print("{0} : 현재 체력은 {1} 입니다".format(self.name, self.hp))
        if self.hp <= 0:
            print("{0} : 해당 유닛이 파괴되었습니다.".format(self.name))

class FlyableUnit:
    def __init__(self, flying_speed):
        self.flying_speed = flying_speed
        
    def fly(self, name, location):
        print("{0} : {1} 방향으로 날아가고 있습니다. [속도 {2}]".format(name, location, self.flying_speed))
        
class FlyableAttackUnit(AttackUnit, FlyableUnit):
    def __init__(self, name, hp, damage, flying_speed):
        AttackUnit.__init__(self, name, hp, 0, damage) # 이런 식으로 기본 메소드 인자에 값을 0으로 지정해서 default값을 전달도 가능하다
        FlyableUnit.__init__(self, flying_speed)
        
    #기존에 지상유닛 즉 Unit 클래스에만 지정되어 있는 move() 함수에 AttackUnit을 통해서 메소드 오버라이딩을 
    #통해 2중으로 메소드 오버라이딩 즉 상위 부모 클래스의 함수에 값을 할당하여 함수를 동작하도록 코드를 작성한 것이다    
    def move(self, location):
        print("[공중 유닛 이동]")
        self.fly(self.name, location)
        
# 유닛 생성 (객체 생성)
vulture = AttackUnit("vulture", 80, 10, 20)
battlecruiser = FlyableAttackUnit("battlecruiser", 500, 25, 3)

vulture.move("11시")
battlecruiser.fly("battlecruiser", "9시")


# 하지만 이런 방식으로 각 유닛별로 즉 객체별로 타입에 따라 사용 가능한 함수가 모두 달라서 타입을 확인하고 move(), fly()와 같은 각각의 함수 구분없이 사용 가능하도록 만들기
# 위해서 메소드 오버라이딩을 적용하여 이런 불편함을 한번에 줄여서 코드 작성이 가능하다
# 그럼 기존에 사용이 불가능하던 공중 유닛에게도 지상 유닛만 사용 가능하던 move() 함수에 값을 전달하여 사용이 가능해진다.
battlecruiser.move("9시")
```
