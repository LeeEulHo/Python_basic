<h2>3. 연산자 및 수식 (+, -, /, %, //, (, ), *, **, >, <, <=, >=, ==, =!, and, or)</h2>

```python
# 더하기 연산자 (addition)
print(32+21)

# 뻴셈 연산자 (subtraction)
print(4239-213)

# 곱셈 연산자 (multiplication)
print(372*2)

# 나눗셈 연산자 (division) 몫(소수점 포함), 나머지, 몫(소수점 제외)
print(321/32)
print(321%32)
print(321//32)

# 제곱 연산자 (square)
print(2**5)

# 일반 수식 연산자 작성의 경우에는 변수를 활용하여 다음과 같이 연산도 가능합 (모든 연산자가 사용 가능하나, 꼭 = 이전에 연산자를 두어아 오류 발생 안하니 주의)
number = 62
number += 32
print(number)
number -= 32
print(number)
number *= 2
print(number)
number /= 2
print(number)
number **= 2
print(number)

# 비교 연산자 (comparison) -> 결과는 True, False로 출력됨 (<=, >=와 같이 사용됩)
print(3<4)
print(3<1)
print(3==2)
print(3>=9)
print(8<=8)
print(8!=8)

# and(&) ,or(|) 연산자
print((3>0) and (2<4)) #-> True + True = True
print((3<0) | (3>9)) #-> False + False = False

# 수학 수식 연산자는 일반 수학 공식과 동일함
print(2+3*4)
print(2+(3*4))
print(2+3/4)
print(2+(3/4))
```

<h2>4. 숫자 관련 처리 함수 (abs, pow, max, min, round, math 라이브러리의 floor, ceil, sqrt, random 함수)</h2>

```python
# abs() : 절댓값
print(abs(-5))

#pow(a,b) : a에 대한 b제곱을 수행함, a**b랑 동일하지만 함수로 사용해야 하는 경우가 있으니 알아 둘 것
print(pow(4,2))

#max(a, b) : a,b 중 최대값을 출력함
print(max(4, 32))

#round() : 입력된 숫자의 소수점을 반올림하여 출력함
print(round(4.18))
print(round(5.99))

#math 라이브러리 import
from math import *

#floor() : 내림(소수점을 모두 버림)
print(floor(4.99))

#ceil() : 올림(소수점이 몇이든 1 높은 숫자로 변경)
print(ceil(16.32))

#sqrt() : 제곱근 값을 출력함(루르값의 결과)
print(sqrt(256))
```

```python
# random 함수 호출 방법
from random import *

# random() : 무작위 난수를 만들어냄. 단, 미 미지정시에는 0.0 ~ 1.0 사이의 값들 중에서 출력됨
print(random()) # 0.0 ~ 1.0 사이의 값
print(random()*10) # 소수점 값에 10을 곱함 (0 ~ 10 미만)
print(int(random()*10)) # int 형으로 변환하여 float 형의 소수점을 모두 날려버림

# random 함수를 활용해서 로또 번호를 랜덤하게 출력한다고 하면 다음과 같이 활용 가능함
list = []
for i in range(6):
    list.append(int(random()*45)+1) #random 값을 0~45이하에서 소수점 없는 정수형으로 만들어져서 list에 들어가도록
list.sort()
print(list)

# randrange() : 범위를 지정한 랜덤값 생성도 가능함
print(randrange(1, 46)) # 1 ~ 46 미만의 값 생성

# randint(a, b) : a~b 범위 내에서 a와 b를 포함한 값 중 랜덤값을 생성함
print(randint(1,45)) # 1 ~ 45의 값들 중 하나를 랜덤하게 출력함

# 따라서, 랜덤함수에는 지정된 값을 제외한 사이의 값 중 출력하는 random, 범위를 지정하여 사이의 값을 출력하는 randrange(), 범위를 지정하면 a,b로 주어진 값까지 포함시켜서 출력되는 randint()
# 가 있기에 추후에 수식이나 알고리즘 또는 DL, ML에서 원하는 수식을 만들어낼 경우에 필요하다면 적합한 함수를 골라서 사용하면 되겠다.
```

<h2>5. 문자열 처리 함수 (슬라이싱, len(), upper(), lower(), isupper(), replace(), index(), find() 등) </h2>

```python
# 변수에 문자열을 지정하여 출력하는 방법
lang = "Korean"
nation = "USA"
print("제 국적은 " + nation + "이며, 사용 가능한 외국어는 " + lang +"입니다.")

# 여러 줄의 문자열을 널어도 정상적으로 출력이 가능하다
sentence = '''
South Korea K-POP is definitly world class
my friend buy a lot of K-POP goods for 10 years
'''
print(sentence)

# 슬라이싱 (문자열 내에서 특정 범위 내의 원하는 문자열만 뽑아서 사용하는 방법)
ID_number = "001231-123234"
print("성별 : " + ID_number[7])
print("연 : " + ID_number[0:2]) #실 자리수 +1 까지 범위로 지정해야 포함됨
print("월 : " + ID_number[2:4])
print("일 : " + ID_number[4:6])
print("생년월일 : " + ID_number[:6]) # :(숫자)로 범위를 정하면 처음부터 지정된 인덱스 이전까지 문자열을 출력함
print("뒤 7자리 : " + ID_number[7:]) # (숫자):로 범위를 정하면 맨 뒷자리부터 지정된 인덱스까지의 문자열을 출력함
```

```python
# 문자열 처리 함수 
python = "Python is AmazingP" # 예시 문자열 준비

#lower() : 소문자로 모든 문자열을 변경
print(python.lower())

#upper() : 대문자로 모든 문자열을 변경
print(python.upper())

#isupper() : 해당 문자열이 대문자가 있는지 -> True/False
print(python[0].isupper())

#len() : 문자열의 길이를 출력함
print(len(python))

#replace("A", "B") : 문자열 A를 찾아서 B로 대체함 (문자, 단어 단위로 가능)
print(python.replace("n", "y"))

#index() : 문자열 내애서 원하는 문자의 위치를 찾을 수 있음
index = python.index("P")
print(index) #P의 문자열 위치를 index에 저장
index = python.index("P", index + 1) #처음 P를 찾은 위치 이후로 P가 있는지 검사하는 방법 -> 없을 경우에는 오류가 발생함
print(index) #두 번째 P의 위치를 출력

#find() : 원하는 문자의 위치를 알려주는 기능을 수행함
print(python.find("JAVA")) #찾을 경우 : 정확한 위치, 없을 경우 : -1

# index()와 find()를 보면 알 수 있듯이 둘 다 특정 문자나 문자열의 위치를 찾는 함수이지만
# index()는 없을 경우 오류를 출력하고 커널 자체를 멈춰버리고
# find()는 없을 경우 -1값을 return 하고 다음 명령을 이어나간다.
# 따라서, 이런 부분에서 차이가 있기 때문에 프로그래밍 하고자 하는 방식에 따라 유념하길 바란다.

#count() : 특정 문자의 반복 횟수를 문자열 내에서 검사하여 출력함
print(python.count("n")) #n의 갯수를 찾아서 출력함
```

<h2>6. 문자열 포맷</h2>

```python
# 기존의 문자열을 + , 를 통해서 이어 붙여서 사용하였다면, 해당 section에서는 다른 방법으로 가능함을 보여준다
print("a" + "b")
print("a", "b")

# %d : 정수형 값을 문자열 자리에 넣는 방법
print("올해는 %d년도 입니다." % 2025)
print("올해는 %d년도 이고, 내년은 %d년입니다" % (2025, 2026))

# %s : 문자열 값을 문자열 자리에 넣는 방법
print("올해 여행지 추천은 %s입니다" % "일본")

# %c : 문자를 문자열 자리에 넣는 방법
print("이번 학기 자료구조 수업의 학점은 %c입니다" % "A")

# {}, .format() -> format() 내에 원하는 데이터를 넣으면 자동으로 {} 내의 자리로 삽입된다.
print("내 올해 나이는 {}살이 됩니다. ".format(25))
print("올해는 {}년도 이고, 내년은 {}년입니다".format(2025, 2026))
print("올해는 {1}년도 이고, 내년은 {0}년입니다".format(2025, 2026)) #{} 내에 뒤에 넣고싶은 format의 문자열 순서를 임의로 변경하여 사용도 가능하다.
print("올해는 {1}년도 이고, 내년은 {1}년입니다".format(2025, 2026)) #{} 내에 같은 format 문자열의 인덱스를 넣으면 중복값이 들어간다
print("나는 {major}전공이며, 원하는 직업은 {job}입니다".format(major="컴퓨터공학과", job="개발자")) #이런 식으로 {}내에 변수를 적용하고 .format 내에서 변수 선언을 하면 적용이 가능하다.
# 해당 부분은 python 3.6 version 이상부터 사용 가능함
major = "Computer Science"
job = "Developer"
print(f"나는 {major}전공이며, 원하는 직업은 {job}입니다") #.format() 대신 변수를 외부에서 선언하고 출력문 앞에 f를 붙여서 {변수}를 적용하면 변수 내용이 문자열 내로 들어간다
```

<h2>8. 데이터 자료형 (List, Dict, Tuple, Set, 데이터 자료형 변경 방법)</h2>
<h3>1) List - index(), append(), insert(), pop(), count(), sort(), reverse(), clear(), extend()</h2>

```python
# List : []로 선언되며 다량의 데이터를 연속적인 공간에 모아서 자료형으로 사용하는 것이 특징이다

score = [90, 89, 52, 18, 99]
print(score)

# 이런 식으로 권장되지는 않지만, 문자열과 정수형 데이터가 함께 리스트에 들어가서 자료형을 이루는 것도 가능은 하다 (int, string, boolean 등 다양한 자료형 가능함)
list_ex = [90, "98", 11, "2"]
print(list_ex)

# index() : List 내부에 원하는 원소의 위치를 찾는 방법
list_ex.index("98")

# append() : 기준 리스트에 새로운 요소를 추가함
list_ex.append(92)
print(list_ex)

# insert(위치, 데이터) : List 내부에 원하는 위치에 삽입하는 방법
list_ex.insert(2, 1923)
print(list_ex)

# pop() : 리스트 맨 뒤에 요소부터 차례로 삭제하고 출력함
print(list_ex.pop())
print(list_ex)

# count() : List 내부에 같은 데이터가 몇개가 존재하는지 확인
list_ex.append(1923)
print(list_ex.count(1923))

# sort() : List 내부의 요소를 정렬
num_list = [2, 9, 43, 4, 3, 8, 12, 5]
num_list.sort()
print(num_list)

# reverse() : List 내부의 요소들의 순서를 앞뒤로 바꿔서 저장함
num_list.reverse()
print(num_list)

# clear() : List 내부의 요소를 모두 삭제함
num_list.clear()
print(num_list)

# extend() : 기존의 List에 다른 List를 확장시켜서 사용도 가능하다
original_list = [1, 2, 3, 4, 5]
extend_list = [6, 7, 8, 9, 10]
original_list.extend(extend_list)
print(original_list)
```

<h3>2) Dict - get(), key in dict_name, del dict_name, keys(), values(), items(), clear() </h3>

```python
# Dict : key와 value 값을 기본으로 구성되는 자료형을 의미한다. 

cabinet = {3:"류현진", 100:"강백호"}
print(cabinet[3])
print(cabinet[100])
#print(cabinet[5]) # 없는 값을 출력할 경우에는 커널 자체라 오류를 출력하고 종료시켜서 다음 명령어는 시행되지 않음

# get() : dict 자료형에서 key값을 활용하여 값을 출력하는 방법
print(cabinet.get(3))
#print(cabinet.get(5)) #값이 없음에도 불구하고 오류를 출력하지 않고 None을 return하고 이어서 다음 명령어를 출력한다
print(cabinet.get(4, "사용가능")) #만약 해당 key값에 해당하는 value 값이 없을 경우에는 출력값을 지정 할 수 있다.

# key in dict_name : dict 자료형 내에 미리 key값에 value가 있는지 여부를 확인이 가능함
print(3 in cabinet) #True
print(5 in cabinet) #False

# key값는 int형이 아닌 string 형도 동일하게 적용 가능함
cabinet = {"A-10":"류현진", "B-100":"유재석"}
print(cabinet["A-10"])
print(cabinet["B-100"])

# Dict 자료형 내부에 새로운 데이터를 추가하는 방법은 dict_name["key"] = value 형으로 적용하면 추가가 가능하다
print(cabinet)
cabinet["C-10"] = "조세호"
cabinet["D-100"] = "김종국"
print(cabinet)

# del dict_name[key] : dict 자료형 내에 key 값에 해당하는 value 값을 삭제가 가능하다
del cabinet["C-10"]
del cabinet["D-100"]
print(cabinet)

# keys() : Dict 자료형 내에 어떤 Key값들이 있는지를 확인이 가능하다
print(cabinet.keys())

# values() : Dict 자료형 내의 어떤 Value값들이 있는지 확인이 가능하다
print(cabinet.values())

# items() : Dict 자료형 내의 모든 데이터들을 Key,Value 형태로 출력함
print(cabinet.items())

# clear() : Dict 자료형 내의 모든 데이터를 삭제함
print(cabinet.clear())
```

```python
# Tuple : 튜플()은 리스트[]와 거의 기능은 비슷하나, 내부의 데이터 원소의 값을 한번 정하면 바꿀 수 없다는 것이 다르다.

# Tuple의 경우에는 ()를 통해 리스트처럼 선언되며 리스트와 동일하게 인덱스값을 통한 활용이 가능하다
menu = ("김밥", "치즈까스")
print(menu[0])
print(menu[1])

# 단, Tuple의 경우 처음 선언 당시 말고 새로운 data값을 넣으려고 할 경우에는 오류가 발생하며 추가가 불가능하다는 것을 알 수 있다.
#menu.add("갈비찜")

# Tuple 자료형을 활용하는 방법으로는 여러 변수를 선언해서 변수명을 통해 튜플 자료형으로 만들어 출력하는 방법도 가능하다
name = "James"
age = 20
hobby = "coding"
print(name, age, hobby)

# 또는, 튜플로 묶어 변수를 선언하고 동일한 구조로 변수 내의 값을 선언하면 다음과 같이 출력도 가능하다.
(name, age, hobby) = ("Harry", 21, "make something")
print(name, age, hobby)
```

<h3> 4) Set - {}, set([List]), add(), remove() 합집합 : &, .intersection() 교집합 : union(), | 차집합 : -, .difference()</h3>

```python
# Set : Set은 Dict 자료형과 동일하게 {}를 통해서 선언하여 사용한다. Dict의 특징인 Key:Value 구조와는 다르게 Set은 Value 값만을 통해서 자료형을 구성한다
# Set 자료형의 특징은 1) 데이터의 중복 선언은 안됨, 2) 순서가 따로 없다는 것이 있다.

# {} : 해당 example에서도 볼 수 있듯이 set 자료형을 선언 시 중복된 데이터가 있어도 출력시에는 중복 데이터는 제거되어 나오는 것을 알 수 있다.
example_set = {1, 2, 3, 3, 4, 5, 5}
print(example_set)

# {}, set([List]) : Set은 수학적 개념으로는 집합이기 때문에 교집합, 합집합, 차집합의 개념도 구현이 가능하다.
team_roaster = {"박지성", "손홍민", "김민재"} #set은 {} 내에 이렇게 선언이 기본이지만
team2_roaster = set(["호날두", "손홍민"]) #[]리스트로 선언하고 후에 set()을 통해 set으로 변환도 가능하다

# &, .intersection() : 교집합에 해당하는 값을 출력
print(team_roaster & team2_roaster) 
print(team_roaster.intersection(team2_roaster))

# |, .union() : 합집합 (중복값을 제외한 모든 합집합의 원소를 출력함)
print(team_roaster | team2_roaster)
print(team_roaster.union())

# -, .difference() : 차집합 (기존 집합에서 교집합을 제외한 원소를 출력함)
print(team_roaster-team2_roaster)
print(team_roaster.difference(team2_roaster))

# add() : 기존 Set 자료형에 추가로 데이터를 삽입
team2_roaster.add("메시")
print(team2_roaster)

# remove() : 기존 Set 자료형 내부의 데이터를 삭제함
team2_roaster.remove("메시")
print(team2_roaster)
```

<h3>자료구조를 변경하는 방법 - list(), tuple(), set() [Dict는 불가]</h3>

```python
# 우선 위에서 언급한 자료형은 List[], Dict{key:value}, Tuple(), Set{}으로 나뉘어 있다. 
# 이런 DataType은 처음 선언할 경우 메모리가 없어지기 전까지 타입이 유지가 되는 것이 아닌 임의로 변경도 가능하다.
# 단, Dict{key:value}는 내부의 key:value 값의 형태 때문에 어떤 자료형으로 변경도 불가능하고 다른 자료형이 Dict 타입으로 변경도 불가능하다.

# Set 자료형을 우선 선언
menu = {"커피", "우유", "주스"}
print(menu, type(menu))

# Set -> List로 변경
menu = list(menu)
print(menu, type(menu))

# List -> Tuple로 변경
menu = tuple(menu)
print(menu, type(menu))

# Tuple -> Set으로 변경
menu = set(menu)
print(menu, type(menu))
```

<h2>9. 조건문 (if, elif, else)<h2>

<h3>if-elif(n개)-else 구조, 조건식 or과 and, 비교 연산자를 활용한 조건식</h3>

```python
# 기본적 모든 programming 언어에 필수적인 조건문 구조로, if(조건), elif(n번째 조건), else(모든 조건 제외 case로 나눠서 
# 조건식에 맞춰서 해당하는 코드를 실행하도록 제어하기 위한 문법으로 사용된다.

'''
기본적인 조건문 if의 구조는 다음과 같이 사용이 가능하다
if(조건):
    실행문
elif(조건): -> n개 사용 가능
    실행문
else: 모든 조건이 아닌 경우
    실행문
'''

# 입력된 값에 따라서 조건문에 따라 답변을 수행하는 조건문의 예시
#weather_condition = input("Today's weather condition : ")
weather_condition = "snow"

#만약 조건이 다중으로 필요할 경우에는 다음과 같이 or로 조건식을 만들 수 있다.
if weather_condition == "rain" or weather_condition == "snow":
    print("take your umbrella")
elif weather_condition == "smoke":
    print("wear a mask today")
else:
    print("clean! have a nice day!")
    
# 비교 조건식을 활용한 온도에 따른 출력문 조건식
#temp = int(input("오늘의 기온은 어떤가요? : "))
temp = 10
if temp >= 30:
    print("오늘은 폭염이니 주의하세요")
elif temp >= 10 and temp < 30:
    print("오늘은 날씨가 좋네요! 좋은 하루 보내세요")
elif temp >= 0 and temp < 10:
#elif 0 <= temp < 10: #과 같이 조건식의 비교 연산자를 바꿔서 사용도 가능함
    print("오늘은 날씨가 다소 쌀쌀하니 외투를 꼭 챙기세요!")
else:
    print("한파로 인해 체감온도가 낮으니 주의하세요!")
```

<h2>10. 반복문 (for(한줄 for문), while, continue, break)</h2>

<h3>1) for - in range(), in List_name

```python
#for 반복문을 이용하여 특정 코드를 반복 실행을 진행한다.

# 반복문이 없다면 기존에는 하나씩 출력문을 작성해야만 한다
# print("대기번호 : 1")
# print("대기번호 : 2")
# print("대기번호 : 3")
# print("대기번호 : 4")

# 하지만, for문을 사용하면 생각보다 간단하게 정리가 가능하다
for waiting_no in range(1, 5): #range의 값만큼 반복하여 waiting_no 변수에 값을 넣음
    print("대기번호 : {0}" .format(waiting_no)) #문자열 format 방식을 이용해서 자동 형변환을 통해서 문자열의 자리에 들어가도록 함
    
# List를 만들어서 List 내부의 값을 하나씩 반복하여 명령어 실행하는 방법도 가능하다.
starbucks = ["아이언맨", "토르", "그루트"]
for customer in starbucks:
    print("{0}, 커피가 준비되었습니다.".format(customer))
```

<h3>2) while -  </h3>

```python
# while 반복문의 경우에는 특정 조건을 만족하기 전까지 지속적으로 반복하는 것이 특징이다. 하지만, 조건이 만약 충족되지 않는경우 즉 break point가 없을 경우에는 무한루프를 반복하며 문제가 발생한다.
# 따라서, while문을 사용할 경우에는 반드시 break point를 지정해서 반복문이 제어가 가능하도록 해야 다른 실행문에 영향을 주지 않는다.

customer = "토르"
index = 5
while index >= 1:
    print("{0}, 커피가 준비 되었습니다. {1}번 남았습니다." .format(customer, index))
    index -= 1
    if index == 0:
        print("커피는 폐기처분되었습니다.")
```
